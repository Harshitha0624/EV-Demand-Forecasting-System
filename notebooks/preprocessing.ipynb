```python
import pandas as pd
```

```python
df = pd.read_csv("../data/ev_charging_data.csv")
df.head()
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df["day_of_week"] = df["date"].dt.dayofweek
df["is_weekend"] = df["day_of_week"].apply(lambda x: 1 if x >= 5 else 0)
df["month"] = df["date"].dt.month
```

```python
df["station_id"] = df["station_id"].astype("category").cat.codes
```

```python
df.head()
```

```python
df.to_csv("../data/ev_charging_data_preprocessed.csv", index=False)
```

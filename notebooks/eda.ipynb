```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("../data/ev_charging_data_preprocessed.csv")
df.head()
```

```python
df.describe()
```

```python
hourly_demand = df.groupby("hour")["energy_kwh"].mean()

plt.figure()
hourly_demand.plot()
plt.xlabel("Hour of Day")
plt.ylabel("Average Energy Demand (kWh)")
plt.title("Average EV Charging Demand by Hour")
plt.show()

```

```python
weekend_demand = df.groupby("is_weekend")["energy_kwh"].mean()

plt.figure()
weekend_demand.plot(kind="bar")
plt.xlabel("Is Weekend (0 = Weekday, 1 = Weekend)")
plt.ylabel("Average Energy Demand (kWh)")
plt.title("Weekday vs Weekend EV Charging Demand")
plt.show()
```

```python
monthly_demand = df.groupby("month")["energy_kwh"].mean()

plt.figure()
monthly_demand.plot()
plt.xlabel("Month")
plt.ylabel("Average Energy Demand (kWh)")
plt.title("Monthly EV Charging Demand Trend")
plt.show()
```

```python
station_demand = df.groupby("station_id")["energy_kwh"].mean()

plt.figure()
station_demand.plot(kind="bar")
plt.xlabel("Station ID")
plt.ylabel("Average Energy Demand (kWh)")
plt.title("Average EV Demand per Charging Station")
plt.show()
```
